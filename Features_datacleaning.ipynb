```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
# Load dataset
df = pd.read_csv("/content/train_features_extended.csv")

# Display first few rows
print("Dataset Loaded Successfully!")
display(df.head())
```

```python
df.info()
```

```python
# Check missing values
print("\nMissing Values in Dataset:\n", df.isnull().sum())
```

```python
# Select numerical features (excluding Image_Path)
numerical_cols = df.columns[1:]

# Apply Min-Max Scaling
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Save the normalized dataset
df.to_csv("supernova_features_normalized.csv", index=False)
print("\nDataset normalized and saved as 'supernova_features_normalized.csv'")
```
